### STEP 7: Histograms

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

### Observation
- Most customers are between 30 and 45 years old.

```python
# Credit Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['credit_score'], bins=30, kde=True)
plt.title('Credit Score Distribution')
plt.show()
```

### Observation
- Credit scores are approximately normally distributed.

```python
# Estimated Salary Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['estimated_salary'], bins=30, kde=True)
plt.title('Estimated Salary Distribution')
plt.show()
```

### Observation
- Salary values are evenly spread among customers.

### STEP 8: Countplots

```python
# Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df)
plt.show()
```

### Observation
Male and female customers are nearly equally distributed.

```python
# Country Distribution
plt.figure(figsize=(7,5))
sns.countplot(x='country', data=df)
plt.show()
```

### Observation
France contains the highest number of customers.

```python
# Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='churn', data=df)
plt.show()
```

### Observation
Majority of customers have not churned

### STEP 9: Boxplots

```python
# Age
plt.figure(figsize=(7,4))
sns.boxplot(x=df['age'])
plt.show()
```

### Observation
Some outliers are present among older customers.

```python
# Balance
plt.figure(figsize=(7,4))
sns.boxplot(x=df['balance'])
plt.show()
```

### Observation
Balance values show a few extreme observations.

```python
# Estimated Salary
plt.figure(figsize=(7,4))
sns.boxplot(x=df['estimated_salary'])
plt.show()
```

### Observation
Salary values are relatively evenly distributed.

### STEP 10: Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')
plt.show()
```

### Observation
Age has positive correlation with churn.
Active members are less likely to churn.
Most variables show weak correlations.

### STEP 11: Scatterplots

```python
# Age vs Balance
plt.figure(figsize=(8,5))
sns.scatterplot(x='age',
                y='balance',
                hue='churn',
                data=df)

plt.show()
```

### Observation
Customers with higher age and balance tend to churn more.

### STEP 12: Pairplot

```python
sns.pairplot(df[['credit_score',
                 'age',
                 'balance',
                 'estimated_salary',
                 'churn']],
             hue='churn')

plt.show()
```

### Observation
Age appears to influence churn more than other variables.

### STEP 13: Relationships and Trends

```python
# Churn by Gender
sns.countplot(x='gender', hue='churn', data=df)
plt.show()
```

```python
# Churn by Country
sns.countplot(x='country', hue='churn', data=df)
plt.show()
```

# Summary of Findings

# Key Findings
- Dataset contains 10,000 records and 11 useful features.
- No missing values or duplicate records were found.
- Majority of customers belong to France.
- Most customers are aged between 30 and 45 years.
- Age is positively correlated with churn, suggesting that older customers tend to have higher churn rates.
- Active members are less likely to leave the bank.
- Germany exhibits relatively higher churn rates.
- Age shows a positive association with churn.
- Credit score has weak correlation with churn.
- –Age and customer activity show noticeable relationships with churn.
